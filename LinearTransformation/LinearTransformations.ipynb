```python
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
file = open('fish.m', 'r')

fish = []

# for line in file:
read = file.read()
m = re.findall('[+-]?[0-9]+\.[0-9]+', read)
m = [float(string) for string in m]

for i in range(0, len(m), 2):
    fish.append([m[i], m[i+1]])
    
fish = np.array(fish)

print(fish)
```

```python
def scaling(x, y, z):
    return np.array([[x, 0, 0],
                     [0, y, 0],
                     [0, 0, z]])

def rotation(theta):
    return np.array([[np.cos(np.deg2rad(theta)), -np.sin(np.deg2rad(theta))],
                     [np.sin(np.deg2rad(theta)), np.cos(np.deg2rad(theta))],
                    ],)

def shear(a):
    return np.array([[1, a],
                     [0, 1]])
```

```python
scaledfish = scaling(2,2)@fish.T

# fig = plt.figure()
plt.scatter(fish[:,0], fish[:,1], color = 'b')
plt.scatter(scaledfish[0], scaledfish[1], color = 'r')
plt.show()
```

```python
rotatedfish = rotation(90)@fish.T

plt.scatter(fish[:,0], fish[:,1], color = 'b')
plt.scatter(rotatedfish[0], rotatedfish[1], color = 'r')
plt.show()
```

```python
shearedfish = shear(2)@fish.T

plt.scatter(fish[:,0], fish[:,1], color = 'b')
plt.scatter(shearedfish[0], shearedfish[1], color = 'r')
plt.show()
```
