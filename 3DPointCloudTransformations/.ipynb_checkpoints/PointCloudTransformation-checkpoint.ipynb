```python
%matplotlib inline
import numpy as np
from numpy import sin, cos, deg2rad as d2r
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as ax3d
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import matplotlib.animation as animation
from IPython.display import HTML, Image
import open3d

```

```python
def homogenous(array):
    return np.vstack((array, np.ones([1, array.shape[1]])))
                                    
def translation(tx, ty, tz):
    return np.array([[1, 0, 0, tx],
                     [0, 1, 0, ty],
                     [0, 0, 1, tz]])

def scaling(sx, sy, sz):
    return np.array([[sx, 0, 0],
                     [0, sy, 0],
                     [0, 0, sz]])
        
def roll(theta):
    return np.array([[1, 0, 0],
                     [0, cos(d2r(theta)), -sin(d2r(theta))],
                     [0, sin(d2r(theta)), cos(d2r(theta))]])
    
def pitch(theta):
    return np.array([[cos(d2r(theta)), 0, sin(d2r(theta))],
                     [0, 1, 0],
                     [-sin(d2r(theta)), 0,  cos(d2r(theta))]])

def yaw(theta):
    return np.array([[cos(d2r(theta)), -sin(d2r(theta)), 0],
                     [sin(d2r(theta)), cos(d2r(theta)), 0],
                     [0, 0, 1]])


'''  Rotation function for all axis  '''
def rotation(xdeg, ydeg, zdeg):
    return yaw(zdeg)@pitch(ydeg)@roll(xdeg)

# def shear(a):
#     return np.array([[1, a],
#                      [0, 1]])
```

```python
# file = 'icosahedron.ply'
file = 'helix.ply'
# file = 'apple.ply'
pcd = open3d.io.read_point_cloud(file)
# convert Open3D.PointCloud to numpy array
xyz = np.asarray(pcd.points)
# print('xyz')
# print(xyz[0])

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter3D(xyz[:, 0], xyz[: ,1], xyz[:, 2], s = 5, c='blue')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
applefile = 'apple.ply'
helixfile = 'helix.ply'
galleonfile = 'galleon.ply'

applepcd = open3d.io.read_point_cloud(applefile)
helixpcd = open3d.io.read_point_cloud(helixfile)
galleonpcd = open3d.io.read_point_cloud(galleonfile)

galleon = np.asarray(galleonpcd.points)
apple = np.asarray(applepcd.points) * 500
helix = np.asarray(helixpcd.points)


fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter3D(apple[:, 0], apple[: ,1], apple[:, 2], s = 5, c='red')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection = '3d')
text = fig.text(0, 1, "TEXT", va='top')  # for debugging
limit = 1000

# scatter = ax.scatter(apple[:, 0], apple[: ,1], apple[:, 2], s = 1, c='red')
# applesc, = ax.plot(apple[:, 0], apple[: ,1], apple[:, 2], '.', markersize=1, color='r')

galleonsc, = ax.plot(galleon[:, 0], galleon[: ,1], galleon[:, 2], '.', markersize=1, color='r')

ax.set_xlim3d([-limit,  limit])
ax.set_xlabel('X')

ax.set_ylim3d([-limit,  limit])
ax.set_ylabel('Y')

ax.set_zlim3d([-limit,  limit])
ax.set_zlabel('Z')
plt.close()

# frames = []
# for i in range(90):
#     frames.append(homogenous(translation(i*100/90,i*100/90,i*100/90)) @
#                    homogenous(rotation(i,0,0) @ scaling(i*90/5000,i*90/5000,i*90/5000)  @ apple.T))

def updatescatter(i, xyz, sc):
    newxyz = homogenous(translation(i*100/90,i*100/90,i*100/90)) @ \
                homogenous(rotation(0, 0, i) @ scaling(i*90/5000,i*90/5000,i*90/5000) @ xyz.T)
    
    # scatter.set_data(frames[i][0], frames[i][1])
    sc.set_data(newxyz[0], newxyz[1])
    sc.set_3d_properties(newxyz[2])
    # scatter.set_3d_properties(frames[i][2])
    text.set_text('frame: %d' % i)  # for debugging

    
ani = animation.FuncAnimation(fig, updatescatter, frames = 90, interval = 33, fargs = [galleon, galleonsc])
HTML(ani.to_html5_video())
# plt.show()
```

```python
ani.save('apple.gif', writer = 'ffmpeg', fps = 30)
```
