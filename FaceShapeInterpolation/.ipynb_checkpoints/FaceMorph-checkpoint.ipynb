```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

from matplotlib.path import Path
from matplotlib.tri import Triangulation
from IPython.display import HTML, Image
from PIL import Image, ImageDraw
from skimage.draw import polygon


```

```python
def landmarks(filename):
    points = 58
    X = np.zeros([2,points])
    # print(x.shape)
    
    file = open(filename)
    
    for i in range(80):
        line = file.readline()
        if (i >= 16) & (i < 16+points):
            split = line.split()
            # print(i-16)
            X[0][i-16] = float(split[2])
            X[1][i-16] = float(split[3])
    # print(X)
    return X
    
    
X1 = landmarks('./01-1m.asf')
img1 = plt.imread('./01-1m.jpg')
X2 = landmarks('./35-1f.asf')
img2 = plt.imread('./35-1f.jpg')

height, width = img1.shape[0], img1.shape[1]

X1[0] *= width
X1[1] *= height

X2[0] *= width
X2[1] *= height

X1 = np.append(X1.T, [[0,0], [0,height], [width,0],[width,height],[0,height/2], [width, height/2], [width/2, 0],[width/2, height]], axis=0)
X2 = np.append(X2.T, [[0,0], [0,height], [width,0],[width,height],[0,height/2], [width, height/2], [width/2, 0],[width/2, height]], axis=0)

X1 = np.hstack([X1, np.ones((X1.shape[0],1))]).astype(np.uint32)
X2 = np.hstack([X2, np.ones((X2.shape[0],1))]).astype(np.uint32)


X1 = X1.T
X2 = X2.T
```

```python
fig = plt.figure(figsize = (10,10))
ax1 = fig.add_subplot(211)
ax1.imshow(img1)
ax1.plot(X1[0], X1[1],'.', c='r')

ax2 = fig.add_subplot(212)
ax2.imshow(img2)
ax2.plot(X2[0], X2[1], '.', c='r')

print(X1.shape)
print(X2.shape)
plt.show()
```

```python
meanX = np.mean([X1,X2], axis = 0).T
print(meanX.shape)

meanX = meanX.T

print(meanX.shape)

meantriangles = Triangulation(meanX[0], meanX[1])

fig = plt.figure(figsize = (10,10))
ax1 = fig.add_subplot(211)
ax1.imshow(img1)
ax1.plot(X1[0], X1[1],'.' , c='r')
ax1.triplot(meantriangles, c='g')

ax2 = fig.add_subplot(212)
ax2.imshow(img2)
ax2.plot(X2[0], X2[1],'.' , c='r')
ax2.triplot(meantriangles, c='g')
plt.show()
```

```python
def EstimateAffine(x, xprime):
    M = np.block([[x, np.zeros((3,3))], [np.zeros((3,3)), x]])
    # print(M)
    b = xprime[:,:2].T.reshape(6,1)
    # print(b)
    
    zeros = np.zeros((1,3))
    zeros[0][-1] = 1
    # print(zeros)
    a = np.linalg.inv(M) @ b
    # print(a)
    A = np.vstack([a.reshape(2,3), zeros])
    # print(A)
    return A
```
